```python
#Theory Questions:-

#1. what is the difference between a function and a method in python?

"""
Functions:-
1. A function is a standalone block of reusable code that performs a specific task. It is defined using the def keyword
2. Functions are not tied to any object and can be called independently/
3.Example:-
 """
def add(a,b):
  return a + b

c = add(3,8)
print(c)#result will be 11

"""
Method:-
1. A method is a function that is associated with an object (usually an instance of a class). It is defined within a class and typically operates on data contained within that object.
2. Methods are called on objects, using dot notation.
3. Example:-
 """
import math

ceil_val = math.ceil(15.25)
print( "Ceiling value of 15.25 is : ", ceil_val)



#2. Explain the concept of function arguments and parameters in python?

"""
Parameters:-A parameter is the variable defined within the parentheses during function definition. Simply they are written when we declare a function.

Arguments:-An argument is a value that is passed to a function when it is called. It might be a variable, value or object passed to a function or method as input. They are written when we are calling the function.

 """
 #Example:-
def sum(a,b):#Here a and b are parameters.
  print(a+b)

sum(5,7) #Here the values 5 & 7 are arguments.



#3. What are the different ways to define and call a function in python?

"""
In python we can define function in various ways:-

1. Basic Function Definition and Call

def greet():
  print("hello how are you")

greet()

2.  Function with Parameters

def add(a,b):
  return a + b

add(20,65)


3.Default Parameters

def greet(name="Ajay"):
    print(f"Hello, {name}!")

greet()
 """


#4. What is the purpose of the 'return' statement in python function?
"""
The return statement is essential for exiting functions and sending values back to the caller.
It allows functions to produce outputs, manage flow control, and handle multiple return values.
If no return is used, the function returns None by default.  """

def add(a, b):
    return a + b

result = add(3, 5)  # The result will be 8
print(result)


#5. What are iterators in Python and how do they differ from iterables?
"""
An iterator is an object that allows traversal through a countable number of values. It is used to iterate over iterable objects like lists, tuples, sets, etc
An iterator is an object that implements the iterator protocol, which consists of two methods:
__iter__(): Returns the iterator object itself.
__next__(): Returns the next value from the iterator
 """
my_list = [5, 7, 3]
print(isinstance(my_list, Iterable))  # Output: True


#6. Explain the concept of generators in Python and how they are defined.

"""
Generators are a simple way to create iterators using functions and the yield keyword instead of returning values.

Use the def Keyword: Like any function, start with the def keyword.
Include yield Statements: Instead of using return, use yield to produce values.
 """
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield the current count and pause the function
        count += 1


#7. What are the advantages of using generators over regular functions?

"""
Memory Efficiency: Generators generate values on the fly, reducing memory usage compared to storing the entire sequence in memory.

Lazy Evaluation: Values are computed only when needed, which can lead to performance improvements, especially with large datasets.

Maintain State: Generators automatically maintain their state between yield statements, making them suitable for iterative processes.

Convenience: Generators simplify code for iterating over large datasets or streams of data without needing to manage state explicitly.
"""

#8.  What is a lambda function in Python and when is it typically used?
"""
 Lambda functions are small, anonymous functions defined using the lambda keyword. They are used for creating small, throwaway functions without the need to formally define a function using def

 Typical Use Cases:-
 1. Short and simple function.
 2. Higher-Order Functions.
 """
add = lambda x, y: x + y
result = add(3, 5)  # Output: 8

#9. Explain the purpose and usage of the `map()` function in Python.
"""
 The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.
 Transformation: Used to transform each item in an iterable by applying the specified function.
 """
number = [1,2,3,4]
squared = map(lambda x: x*x, number)
print(list(squared))

#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
"""
map()
Purpose: Applies a function to every item in an iterable (like a list) and returns a map object (which can be converted to a list).
Usage: Typically used when you want to transform the elements of an iterable.
Example:"""
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))  # Output: [1, 4, 9, 16]


"""filter()
Purpose: Filters elements from an iterable based on a function that returns True or False. It returns a filter object, which can also be converted to a list.
Usage: Used when you want to retain only those elements that meet a specific condition.
Example:"""
numbers = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x % 2 == 0, numbers))  # Output: [2, 4]

 """
 reduce()
Purpose: Applies a function of two arguments cumulatively to the items of an iterable, reducing the iterable to a single value. It is part of the functools module and needs to be imported.
Usage: Typically used when you want to compute a single result from a collection of items (like a sum or product).
Example:"""
from functools import reduce

numbers = [1, 2, 3, 4]
sum_result = reduce(lambda x, y: x + y, numbers)  # Output: 10



#Practical Questions:

#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total


num_list = [1, 2, 3, 4, 5, 6, 7, 8, 45, 98]
result = sum_of_even_numbers(num_list)
print(result) #output: 118


#2. Create a Python function that accepts a string and returns the reverse of that string.

def string_reverse(s):
    return s[::-1]

input_string = "physics_wallah"
reversed_string = string_reverse(input_string)
print(reversed_string)  # Output: "hallaw_scisyhp"


#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square(numbers):
    squared_list = []
    for num in numbers:
        squared_list.append(num ** 2)
    return squared_list

input_list = [6, 5, 4, 3, 2]
squared_result = square(input_list)
print(squared_result)

#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

primes = [num for num in range(1, 201) if prime(num)]
print(primes)


#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci(terms):
    a, b = 0, 1
    for _ in range(terms):
        yield a
        a, b = b, a + b

for number in fibonacci(15):
    print(number)



#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(7):
    print(power)



#7. Implement a generator function that reads a file line by line and yields each line as a string

#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 'banana'), (3, 'apple'), (2, 'orange'), (4, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)

#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 32, 71, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)


#10. Create a Python program that uses `filter()` to remove all the vowels from a given string

def vowels_remove(a):
    vowels = 'aeiouAEIOU'
    for vowel in vowels:
        a= a.replace(vowel, '')
    return a


a = "Hello Everyone"
result = vowels_remove(a)
print(result)

#Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
```
